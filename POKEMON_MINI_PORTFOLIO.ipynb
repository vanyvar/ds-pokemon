Number of strongest Legendary Pokemon per Generation (total legendary per generation and 3 strongest per generation)

```python
import pandas as pd
#read file
data = pd.read_csv('Pokemon.csv')
#rename '#' to be ID and set as index
data = data.rename(columns={"#":"ID"})
data.set_index("ID", inplace=True)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
#check data null
data.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#sum legendary per generation
sum_legendary = data[data['Legendary'] ==  True].groupby('Generation')[['Legendary']].count()

#visualization use barplot
colors = ['#588E47', '#37642A']
plt.figure(figsize=(10,6))
sns.barplot(x='Generation', y='Legendary', data=sum_legendary, palette=colors)
plt.title('Total Legendary Character per Generation')
plt.xlabel('Generation')
plt.ylabel('Total Legendary Character')
plt.grid(axis='y')

plt.show()
```

5 Strongest Legendaries and Non-Legendaries from 6 Generations

> Determining a character's highest score as a legend or not.


> Will having 2 types of powers make a character stronger?



```python
#top five strongest character legendary
top_legendary = data[data['Legendary'] == True ].sort_values(by='Total', ascending=False)[['Name','Generation','Legendary','Total']].head(5)
top_legendary
```

```python
#top five strongest character non legendary
top_nonlegendary = data[data['Legendary'] == False].sort_values(by='Total', ascending=False)[['Name','Generation','Legendary','Total']].head(5)
top_nonlegendary
```

```python
#total power relationship between legendary or non legendary characters, fetch 400 data
compare_char = data.sort_values(by='Total', ascending = False)[['Generation','Legendary','Total']].head(400)

#visualization with scatterplot
plt.figure(figsize=(8,6))
colors = ['#FFEF00', '#0B6623']
sns.scatterplot(x='Generation', y='Total', hue='Legendary', palette=colors, data=compare_char)
plt.title('Does the strongest have to be owned by a legendary character?')
plt.xlabel('Generation')
plt.ylabel('Total')
plt.grid(axis='y')
plt.show(), compare_char.head(20)
```

```python
#A character that has 2 types will be the strongest or not
type_char = data.sort_values(by='Total', ascending = False)[['Name','Type 1','Type 2','Generation','Legendary','Total']].head(10)
type_char
```

Which character is the strongest, fastest and has good defense

```python
#strongest charachter based on attack, def, speed
att_char= data.sort_values(by='Attack', ascending=False)[['Name','Generation','Legendary','Attack']].head(5)
speed_char= data.sort_values(by='Speed', ascending=False)[['Name','Generation','Legendary','Speed']].head(5)
def_char= data.sort_values(by='Defense', ascending=False)[['Name','Generation','Legendary','Defense']].head(5)
att_char, speed_char, def_char

#visualization use barplot
fig, ax=plt.subplots(1,3,figsize=(15,5))
sns.barplot(x='Name', y='Attack', data=att_char, palette=colors, ax=ax[0])
sns.barplot(x='Name', y='Speed', data=speed_char, palette=colors, ax=ax[1])
sns.barplot(x='Name', y='Defense', data=def_char, palette=colors, ax=ax[2])
#set title
ax[0].set_title('Strongest Character based on Attack')
ax[1].set_title('Strongest Character based on Speed')
ax[2].set_title('Strongest Character based on Defense')
#name rotation
ax[0].tick_params(axis='x', rotation=45)
ax[1].tick_params(axis='x', rotation=45)
ax[2].tick_params(axis='x', rotation=45)
plt.tight_layout()
```

Average type possessed by Legend Characters

```python
#filter with legendary pokemon
character_type = data[data['Legendary'] == True]
#join column type 1 and type 2
join_type = pd.concat([character_type['Type 1'], character_type['Type 2']])
#count values
count_type = join_type.value_counts().astype(int)

#visualization with barplot
plt.figure(figsize=(10,6))
sns.barplot(x=count_type.index, y=count_type.values, palette='viridis')
plt.title('Legendary Pokemon Type Frequency')
plt.xlabel('Type')
plt.ylabel('Frequency of occurrence')
plt.xticks(rotation=45)
plt.show()
```
